# Annual Percentage Rate of Return (APR)

```python
stock_change_apr1 = stock_change1 * 252 * 100    
#stock_change_apr1.tail()
```

```python
stocks_start='2015-01-02'
stocks_end='2020-03-05'
end_of_last_year = '2019-12-31'

def get(list1, startdate, enddate):
    def data(list1):
        return (pdr.get_data_yahoo(list1, start=startdate, end=enddate))
    datas = map(data, list1)
    return(pd.concat(datas, keys=list1, names=['Ticker', 'Date']))
               
all_data = get(list1, stocks_start, stocks_end)
```

```python
adj_open = all_data[['Adj Close']].reset_index()
adj_open_1 = adj_open[adj_open['Date']==stocks_start]
adj_open_1.set_index('Ticker', inplace=True)
```

```python
adj_close = all_data[['Adj Close']].reset_index()
adj_close_start = adj_close[adj_close['Date']==end_of_last_year]
adj_close_latest = adj_close[adj_close['Date']==stocks_end]
adj_close_latest.set_index('Ticker', inplace=True)
```

```python
position=5000

adj_close_latest['Unit_Cost']=adj_open_1['Adj Close']
adj_close_latest['Quantity']=position/adj_close_latest['Unit_Cost']
adj_close_latest['Cost_Basis']=position
adj_close_latest['Total_Pos']=adj_close_latest['Quantity']*adj_close_latest['Adj Close']
adj_close_latest['Return_$']=adj_close_latest['Total_Pos']-adj_close_latest['Cost_Basis']
adj_close_latest['Return_%']=adj_close_latest['Adj Close'] / adj_close_latest['Unit_Cost'] - 1

#adj_close_latest
```
